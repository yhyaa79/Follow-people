```python
name_video = "inpot_video2.mov"
```

```python
import cv2

# لیست برای ذخیره نقاط
points = []
door_labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']
door_points = {}
current_door = 0  # شمارش درها

# تابع برای ثبت نقاط کلیک شده
def select_points(event, x, y, flags, param):
    global current_door, frame
    if event == cv2.EVENT_LBUTTONDOWN:  # وقتی کلیک چپ ماوس انجام می‌شود
        points.append((x, y))
        print(f"Point selected for door {door_labels[current_door]}: {(x, y)}")

        # رسم دایره قرمز در محل کلیک
        cv2.circle(frame, (x, y), 5, (0, 0, 255), -1)

        # رسم خطوط بین نقاط
        if len(points) > 1:
            cv2.line(frame, points[-2], points[-1], (0, 255, 0), 2)

# بارگذاری ویدئو
cap = cv2.VideoCapture(name_video)

# خواندن اولین فریم ویدئو
ret, frame = cap.read()

# تنظیم نام پنجره و تنظیمات ماوس
cv2.namedWindow("Frame")
cv2.setMouseCallback("Frame", select_points)

while True:
    # نمایش فریم
    cv2.imshow("Frame", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):  # با فشار دادن 'q'، مجموعه فعلی بسته می‌شود
        door_points[door_labels[current_door]] = points.copy()  # ذخیره نقاط با برچسب
        points.clear()  # خالی کردن لیست برای درب بعدی
        print(f"{door_labels[current_door]} points saved.")
        current_door += 1  # افزایش شمارش درها

        # اگر تعداد درهای مورد نظر تمام شد، حلقه متوقف شود
        if current_door == len(door_labels):
            break

    elif key == ord('r'):  # با فشار دادن 'r'، نقاط درب جاری پاک می‌شوند
        points.clear()  # پاک کردن لیست نقاط
        ret, frame = cap.read()  # بازنشانی فریم اصلی
        print(f"Points for door {door_labels[current_door]} cleared.")

    # خروج با فشار دادن کلید 'ESC'
    if key == 27:  # کلید ESC
        break

cap.release()
cv2.destroyAllWindows()

# نمایش نقاط انتخاب شده برای هر در
print("Selected Points for each door:")
for label, points in door_points.items():
    print(f"{label}: {points}")

# ذخیره نقاط به فایل یا استفاده در پردازش‌های بعدی
# به عنوان مثال ذخیره در فایل
with open("points.txt", "w") as file:
    for label, points in door_points.items():
        file.write(f"{label}:\n")
        for point in points:
            file.write(f"{point[0]},{point[1]}\n")
        file.write("\n")
```

```python
import cv2

# لیست‌ها برای ذخیره مختصات نقاط کلیک‌شده
points = []
area1 = []
area2 = []

# تابعی که برای کلیک ماوس استفاده می‌شود
def select_point(event, x, y, flags, param):
    global points, area1, area2
    if event == cv2.EVENT_LBUTTONDOWN:  # وقتی که روی تصویر کلیک می‌کنید
        points.append((x, y))
        if len(points) == 4:
            if not area1:
                area1 = points.copy()
                print("Area 1 selected:", area1)
            elif not area2:
                area2 = points.copy()
                print("Area 2 selected:", area2)
            points = []

# ویدیو را باز می‌کند
cap = cv2.VideoCapture(name_video)
ret, frame = cap.read()

if ret:
    cv2.namedWindow("Frame")
    cv2.setMouseCallback("Frame", select_point)

    while True:
        frame_copy = frame.copy()

        # رسم نقاط کلیک‌شده فعلی روی تصویر
        for pt in points:
            cv2.circle(frame_copy, pt, 5, (0, 255, 0), -1)

        cv2.imshow("Frame", frame_copy)
        
        key = cv2.waitKey(1) & 0xFF
        if key == 27:  # اگر کلید ESC فشار داده شود، از حلقه خارج شود
            break
        elif key == ord('r'):  # اگر کلید 'r' فشار داده شود، لیست نقاط ریست شود
            points = []
            area1 = []
            area2 = []
            print("Reset points")

cap.release()
cv2.destroyAllWindows()

# چاپ مختصات نهایی
print("Area 1:", area1)
print("Area 2:", area2)
```
